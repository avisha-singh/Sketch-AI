```python
import shutil
import os
```

```python


def copy_and_rename_file(source_file_path, destination_directory, new_file_name):
    # Construct the full path for the new file
    new_file_path = os.path.join(destination_directory, new_file_name)
    # Copy the file
    shutil.copy2(source_file_path, new_file_path)
    print(f"File copied to {new_file_path}")


```

```python
def get_file_list(directory):
    # List all files in the directory
    file_list = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    return file_list
```

```python
file_list=get_file_list("./image")
```

```python
for i in file_list:
    source_file_path = f"./sketch2/{i[:-4]}_01.jpg"  # Replace with your source file path
    destination_directory = "./sketch"    # Replace with your destination directory
    new_file_name = i                       # Replace with your new file name

    copy_and_rename_file(source_file_path, destination_directory, new_file_name)
```

```python
# Usage example
source_file_path = "./source_directory/old_name.txt"  # Replace with your source file path
destination_directory = "./destination_directory"    # Replace with your destination directory
new_file_name = "new_name.txt"                        # Replace with your new file name

copy_and_rename_file(source_file_path, destination_directory, new_file_name)
```
